```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import math as mt
from math import e
from tqdm import tqdm
```

$$  I = \iiint e^{\sqrt{x^2+y^2+z^2}} \,dx\,dy\,dz = 4 \pi (e-2)$$

```python
def circulo( N, R ):
    
    X = np.zeros(N)
    Y = np.zeros_like(X)
    Z = np.zeros_like(X)
    
    
    for i in range(N):
        
        u = np.random.rand()
        r = R*u**(1/3)
        theta = np.random.uniform(0,np.pi*2)
        phi = np.random.uniform(0,np.pi)
        
        X[i] = r*np.sin(theta)*np.cos(phi)
        Y[i] = r*np.sin(theta)*np.sin(phi)
        Z[i] = r*np.cos(theta)
    
    
    return X,Y,Z
```

```python
X,Y,Z = circulo(10000, 1)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, projection = '3d')

#configurar lmites de los ejes
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)

ax.view_init(20, 100)

ax.scatter(X,Y,Z,color='b')
```

```python
def funcion(x,y,z):
    return np.exp(np.sqrt(x**2+y**2+z**2))
```

```python
def puntos(N,R):
    
    suma=0
    
    for i in tqdm(range (N)):
        
        u = np.random.rand()
        r = R*u**(1/3)
        i=np.random.uniform(-1,1)
        theta = mt.acos(i)
        phi = np.random.uniform(0,np.pi)
        
        X = r*np.sin(theta)*np.cos(phi)
        Y = r*np.sin(theta)*np.sin(phi)
        Z = r*np.cos(theta)   
    

        suma += funcion(X,Y,Z)
        
    prom= suma/N
    I= prom*(4/3)*np.pi
    return I       
        
```

```python
I= puntos(1000000,1)
print("Valor aproximado: ", I)
print( "Valor real: ", 4*np.pi*(e-2))
```
